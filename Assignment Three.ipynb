### 1. Read the stock information for Amazon using Pandas' data reader.

```python
import numpy as np
import pandas as pd
from pandas_datareader import data, wb
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Plot a time series of the "volume" from January 1, 2022 to today.

```python
# I need to create a start and end time for the line chart we will be creating
start = datetime.datetime(2022,1,1)
end = datetime.datetime(2022,2,15)
```

```python
# Now, I am calling on the Amazon and specifying the use of ggplot
AMZN = data.DataReader("AMZN", 'yahoo', start, end)
plt.style.use('ggplot')
AMZN
```

```python
AM = AMZN.reset_index()
```

```python
AM.plot.line(x='Date',y='Volume',figsize=(15,4),lw=2,title='Amazon Volume 2022')
plt.show()
```

#### Plot a bar chart of the "opening" for the month of December 2021. 

```python
# For the secord section of part a, I need to create a new start and end time
start2 = start = datetime.datetime(2021,12,1)
end2 = datetime.datetime(2021,12,31)
AMZ = data.DataReader("AMZN", 'yahoo', start2, end2)
```

```python
AM2 = AMZ.reset_index()
AM2
```

```python
AM2.plot.bar(x='Date', y='Open',figsize=(15,6),lw=2,title='Amazon December Openings')
plt.show()
```

### 2. Read the stock information for Microsoft using Pandas' data reader.

```python
plt.style.use('bmh')
```

#### Plot a time series of the "volume" from January 1, 2022 to today.

```python
start3 = datetime.datetime(2022,1,1)
end3 = datetime.datetime(2022,2,15)
```

```python
Micro = data.DataReader("MSFT",'yahoo',start3,end3)
Micro
```

```python
MSFT = Micro.reset_index()
MSFT
```

```python
MSFT.plot.line(x='Date', y='Volume', figsize=(20,6),lw=2,title='Microsoft Volume 2022')
plt.show()
```

#### Plot a density chart of the 'High' and 'Low' values for the month of January 2022.

```python
start4 = datetime.datetime(2022,1,1)
end4 = datetime.datetime(2022,1,31)
```

```python
MSF = data.DataReader("MSFT", 'yahoo', start4, end4)
MSF
```

```python
MS = MSF.reset_index()
MS
```

```python
MS[['High','Low']].plot.density(figsize=(20,6),lw=3,title="Microsoft January Density")
plt.show()
```
